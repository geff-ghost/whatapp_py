## Data Handling Basics Part 3: Data Cleaning with Pandas  

Common problems:  
- Missing values
- Duplicate rows
- Wrong data types
- Extra spaces / inconsistent values

```python
# Import pandas 
import pandas as pd
```

```python
# Detect Missing Values
data = {
    "Name": ['Amit', 'Deepak', 'Ravi', 'Neha'],
    "Marks": [78, None, 90, 82]
}

df = pd.DataFrame(data)
print(df)
```

NaN means missing value.

**Check Missing Values**

```python
# Check Missing Values
print(df.isnull())

print(f"Count missing values: {df.isnull().sum()}")
```

**Remove Missing Values**

```python
# Remove Missing Values
df = df.dropna()
print(df)
```

This removes rows with missing data.

```python
df
```

**Fill Missing Values**  

Instead of removing rows:

```python
df
```

```python
data = pd.DataFrame({
    'Name': ['Amit', 'Deepak', 'Ravi', 'Neha'],
    'Marks': [78, 85, 85, 90]
})
```

```python
# fill Missing values with 0
df['Marks'] = df['Marks'].fillna(0)
```

```python
# Or fill with mean
df['Marks'] = df['Marks'].fillna(df['Marks'].mean())
```

```python
df
```

**Remove Duplicate Rows**

```python
df = pd.DataFrame({
    "Name": ['Amit', 'Deepak', 'Deepak', 'Ravi'],
    "Marks": [78, 85, 85, 90]
})

print(df)
```

```python
# Remove duplicates: 
# df.drop_duplicates() drops a row if every value in that row matches every value in another row
df = df.drop_duplicates()
print(df)
```

**Rename Columns**

```python
df.rename(columns={"Marks":"Score"}, inplace=True)
print(df)
```

**Change Data Types**

```python
df["Score"] = df["Score"].astype(float)

df
```

**Real Example - Clean Student Dataset**

```python
df = pd.read_csv('students.csv')
df = df.drop_duplicates()
df["Marks"] = df["Marks"].fillna(df["Marks"].mean())
print(df.head())
```

## Sales Data Analysis Project  

Analyze sales data to answers:
- Total sales
- Best-selling product
- Average sales
- Filter high-performing products

**1. Sampled Dataset (sales.csv)**  

**2. Load Data**

```python
df = pd.read_csv('sales.csv')

print(df.head())
```

**3. Basic Analysis**

```python
# Total Sales
print("Total Sales:", df['Sales'].sum())

# Average Sales
print("Average Sales:", df['Sales'].mean())

# Maximum Sale
print("Max Sale:", df['Sales'].max())
```

**4. Group by**

```python
# Group by product
grouped = df.groupby("Product")['Sales'].sum()
print(grouped)
```

**5. Find Best-Selling Product**

```python
print("Best Product:", grouped.idxmax())
```

**6. Filter High Sales**

```python
high_sales = df[df['Sales'] > 150]
print(high_sales)
```
