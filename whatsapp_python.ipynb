# Public Variables (Normal Variables)
Accessible from anywhere.

```python
class Student:
    def __init__(self):
        self.name = "Adam"

s = Student()
print(s.name)
```

# Protected Variables (_variable)
Single underscore -> 'internal use' warning.

```python
class Student:
    def __init__(self):
        self._marks = 80

s = Student()
print(s._marks)
```

# Private Variables (__variable)
Double underscore -> hidden from outside.

```python
class Student:
    def __init__(self):
        self.__marks = 90

s = Student()
```

# Access Private Data using Methods (Getter and Setter)
Getter -> read value

```python
class Student:
    def __init__(self):
        self.__marks = 90
    def get_marks(self) -> int:
        return self.__marks

s = Student()
print(s.get_marks())
```

Setter -> update value safely

```python
class Student:
    def __init__(self):
        self.__marks = 90
    def set_marks(self, marks):
        if marks >= 0:
            self.__marks = marks
    def get_marks(self):
        return self.__marks

s = Student()
s.set_marks(85)
print(s.get_marks())
# Now data is controlled
```

# Inheritance (Reusing Code Between Classes)
## What is Inheritance
Inheritance = one class uses properties of another class.

```python
# Parent Class (Base Class)
class Animal:
    def eat(self):
        print("Animal eats")

# Child Class (Derived Class)
class Dog(Animal):
    def bark(self):
        print("Dog barks")

d = Dog()
d.eat()
d.bark()
```

```python
# Method Overriding (Very Important)
class Animal:
    def sound(self):
        print("Animal makes sound")
class Dog(Animal):
    def sound(self):
        print("Dog barks")

d = Dog()
d.sound()

f = Animal()
f.sound()
```

```python
# Using super() (Call Parent Method)
class Animal:
    def eat(self):
        print("Animal eats")
class Dog(Animal):
    def eat(self):
        super().eat()
        print("Dog eats bones")

d = Dog()
d.eat()
```

```python
# Multiple Inheritance
class A:
    pass
class B:
    pass
class C(A, B):
    pass
```

# What is Polymorphism
Poly = many, Morphism = forms

## Example 1. Method Overriding (Runtime Polymorphism)

```python
class Animal:
    def sound(self):
        print("Animal makes sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

class Cat(Animal):
    def sound(self):
        print("Cat meows")

animals = [Dog(), Cat()]
for animal in animals:
    # Same method, Different outputs
    animal.sound()
```

## Example 2. Same Function, Different Objects

```python
class Car:
    def moves(self):
        print("Car moves")

class Plane:
    def moves(self):
        print("Plane flies")

def start_movement(vehicle):
    vehicle.moves()

# Function doesn't care about object type. This is polymorphism
start_movement(Car())
start_movement(Plane())
```

## Example 3. Built-in Polymorphism

```python
print(len("Python"))
print(len([1, 2, 3]))
# Same function works for string and list
```

**Types of Polymorphism**
1. Method Overiding Child class changes parent method.
2. Duch Typing(Python Style) "If it walks like a duck and quacks like a duck, it's a duck." Python doesn't check type - it checks behavior.

```python
class CreditCard:
    def pay(self, amount):
        print(f"Paid ${amount} using Credit Card.")

class UPI:
    def pay(self, amount):
        print(f"Paid ${amount} using UPI.")

def process_payment(method, amount):
    method.pay(amount)

process_payment(CreditCard(), 1000)
process_payment(UPI(), 300)
# same method pay()
# different implementations
```

Abstraction = show only important details and hide implementation.
- Hide how something works
- Show what it does

Car driving

- Engine mechanism is hidden  

**Why Abstraction is Important**  

- Reduce complexity
- Improve security
- Focus on essential features
- Cleaner system design

**Abstract Class (Blueprint Only)**  

Abstract class cannot create objects directly.

```python
from abc import ABC, abstractmethod
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
```

You cannot do: `s = Shape()` # Error  

Why? Because method not defined.

**Child Class Must Implement Abstract Method**

```python
from abc import ABC, abstractmethod
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
class Circle(Shape):
    def __init__(self, r):
        self.r = r
    def area(self):
        return 3.14*self.r*self.r

c = Circle(5)
print(c.area())
```

**Another Example. Payment System**

```python
from abc import ABC, abstractmethod
class Payment(ABC):
    @abstractmethod
    def pay(self, amount):
        pass
class UPI(Payment):
    def pay(self, amount):
        print(f"Paid: ${amount} using UPI")
class CreditCard(Payment):
    def pay(self, amount):
        print(f"Paid: ${amount} using CreditCard")

p = CreditCard()
p.pay(1000)
```

User only sees pay() - implementation hidden.

**Why You Need the Abstract Method**

The abstract method acts as a contract or blueprint that forces all child classes to implement certain methods. Here's why it's useful:

```python
class Shape:  # No abstract method
    def area(self):
        pass  # Empty method

class Circle(Shape):
    def __init__(self, r):
        self.r = r
    # Forgot to implement area() - no error!

circle = Circle(5)
print(circle.area())  # Returns None (silent failure)
```

**Project Goal:**  

Build a simple bank system that:
- Creates accounts
- Deposits money
- Withdraws money
- Shows balances
- Supports different account types

**Step 1: Abstract Base Class**

```python
from abc import ABC, abstractclassmethod
class BankAccount(ABC):
    def __init__(self, name, balance):
        self.name = name
        self.__balance = balance # encapsulated
    @abstractmethod
    def account_type(self):
        pass
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}")
    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn: ${amount}")
        else:
            print("Insufficient balance")
    def get_balance(self):
        return self.__balance
```

**Step 2: Savings Account**

```python
class SavingsAccount(BankAccount):
    def account_type(self):
        return "Savings Account"
```

**Step 3: Current Account**

```python
class CurrentAccount(BankAccount):
    def account_type(self):
        return "Current Account"
```

**Step 4: Test the System**

```python
acc1 = SavingsAccount("Deepak", 1000)
acc2 = CurrentAccount("Amit", 5000)

print(acc1.account_type())
acc1.deposit(500)
acc1.withdraw(300)

print(f"Balance: ${acc1.get_balance()}")
print("\n", acc2.account_type())

acc2.withdraw(6000)
print(f"Balance: ${acc2.get_balance()}")
```

**Concepts Used**  
- Abstraction: BankAccount abstract class
- Encapsulation: __balance private variable
- Inheritance: Savings & Current inherit BankAccount
- Polymorphism: account_type() method

## Project 1: To-Do List App  

**Project Goal**  

Build a command-line application that can:  

- Add tasks
- View tasks
- Delete tasks
- Save tasks for future use  

**Basic Structure**  

- List -> store tasks
- Functions -> separate features
- Loop -> keep program running
- File Handling -> save & load data

```python
# To-Do List App
tasks = []

def show_tasks():
    if not tasks:
        print("No tasks available")
    else:
        print("\nYour Tasks:")
        for i, task in enumerate(tasks):
            print(f"{i+1}. {task}")

def add_task():
    task = input("Enter tasks: ")
    tasks.append(task)
    print("Task added")

def delete_task():
    show_tasks()
    try:
        index = int(input("Enter task number to delete: ")) -1
        if 0 <= index < len(tasks):
            removed = tasks.pop(index)
            print("Removed:", removed)
        else:
            print("Invalid index")
    except:
        print("Enter valid number")

def save_tasks():
    with open("tasks.txt", "w") as file:
        for task in tasks:
            file.write(task + "\n")

def load_tasks():
    try:
        with open("tasks.txt", "r") as file:
            for line in file:
                tasks.append(line.strip())
    except:
        pass
```

## Project 2: Expense Tracker  

**Project Goal**  

Build a simple application that:  

- Adds daily expenses
- Views all expenses
- Calculates total spending
- Saves data for future use  

**Basic Structure**  

- List -> store expenses
- Dictionary -> store expense details (name + amount)
- Functions -> organize features
- File Handling (CSV) -> save & load expenses

```python
import csv
expenses = []

def add_expense():
    name = input("Enter expense name: ")
    amount = float(input("Enter amount: "))
    expenses.append({"name": name, "amount": amount})
    print("Expense added")

def show_expenses():
    if not expenses:
        print("No expenses found")
    else:
        print("\nExpenses: ")
        for i, exp in enumerate(expenses):
            print(f"{i+1}. {exp['name']} - ${exp['amount']}")

def total_expense():
    total = sum(exp['amount'] for exp in expenses)
    print("Total Expense:", total)

def save_expenses():
    with open("expenses.csv", "w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["Name", "Amount"])
        for exp in expenses:
            writer.writerow([exp['name'], exp['amount']])

def load_expenses():
    try:
        with open('expenses.csv', 'r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                expenses.append({
                    'name': row["Name"],
                    'amount': float(row["Amount"])
                })
    except:
        pass

# Load previous data
load_expenses()
```
