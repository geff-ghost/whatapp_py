## Object-Oriented Programming Part 2

# Encapsulation
Encapsulation = protecting data inside a class

# Public Variables (Normal Variables)
Accessible from anywhere.

```python
class Student:
    def __init__(self):
        self.name = "Adam"

s = Student()
print(s.name)
```

```python
# Problem -> anyone can change data.
s.name = "Amit"
print(s.name)
```

# Protected Variables (_variable)
Single underscore -> 'internal use' warning.

```python
class Student:
    def __init__(self):
        self._marks = 80

s = Student()
print(s._marks)
```

# Private Variables (__variable)
Double underscore -> hidden from outside.

```python
class Student:
    def __init__(self):
        self.__marks = 90

s = Student()
print(s.__marks)
```

# Access Private Data using Methods (Getter and Setter)
Getter -> read value

```python
class Student:
    def __init__(self):
        self.__marks = 90
    def get_marks(self) -> int:
        return self.__marks

s = Student()
print(s.get_marks())
```

Setter -> update value safely

```python
class Student:
    def __init__(self):
        self.__marks = 90
    def set_marks(self, marks):
        if marks >= 0:
            self.__marks = marks
    def get_marks(self):
        return self.__marks

s = Student()
s.set_marks(85)
print(s.get_marks())
# Now data is controlled
```

# Real Example. Bank Account Protection.

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
    def get_balance(self):
        return self.__balance

acc = BankAccount(1000)
acc.deposit(500)
print(acc.get_balance())
```

# Practice Task Solutions
## Task 1. Create class Employee with private salary

```python
class Employee:
    def __init__(self, salary):
        self.__salary = salary # private variable

    def get_salary(self):
        return self.__salary

emp = Employee(50000)
print("Salary:", emp.get_salary())
```

## Task 2. Create class Bank with private balance and deposit method

```python
class Bank:
    def __init__(self, balance: int):
        self.__balance = balance
    
    def deposit(self, amount: float):
        if amount > 0:
            self.__balance += amount

    def get_balance(self):
        return self.__balance

b = Bank(1000)
b.deposit(500)
print("Balance:", b.get_balance())
```

## Task 3. Create class Car with private spedd and update method

```python
class Car:
    def __init__(self, speed):
        self.__speed = speed
    
    def update_speed(self, speed):
        if speed >= 0:
            self.__speed = speed
    
    def show_speed(self):
        print("Speed:", self.__speed)

c = Car(80)
c.update_speed(90)
c.show_speed()
```

## Task 5. Prevent negative values using setter 

```python
class Product:
    def __init__(self, price):
        self.__price = 0
        self.set_price(price)

    def set_price(self, price):
        if price >= 0:
            self.__price = price
        else:
            print("Price cannot be negative")

    def get_price(self):
        return self.__price

p = Product(100)
p.set_price(-20) # invalid
print("Price:", p.get_price())
```

# Inheritance (Reusing Code Between Classes)
## What is Inheritance
Inheritance = one class uses properties of another class.

```python
# Parent Class (Base Class)
class Animal:
    def eat(self):
        print("Animal eats")

# Child Class (Derived Class)
class Dog(Animal):
    def bark(self):
        print("Dog barks")

d = Dog()
d.eat()
d.bark()
```

```python
# Method Overriding (Very Important)
class Animal:
    def sound(self):
        print("Animal makes sound")
class Dog(Animal):
    def sound(self):
        print("Dog barks")

d = Dog()
d.sound()

f = Animal()
f.sound()
```

```python
# Using super() (Call Parent Method)
class Animal:
    def eat(self):
        print("Animal eats")
class Dog(Animal):
    def eat(self):
        super().eat()
        print("Dog eats bones")

d = Dog()
d.eat()
```

```python
# Multiple Inheritance
class A:
    pass
class B:
    pass
class C(A, B):
    pass
```
