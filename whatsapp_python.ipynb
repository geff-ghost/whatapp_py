# Encapsulation
Encapsulation = protecting data inside a class

# Public Variables (Normal Variables)
Accessible from anywhere.

```python
class Student:
    def __init__(self):
        self.name = "Adam"

s = Student()
print(s.name)
```

```python
# Problem -> anyone can change data.
s.name = "Amit"
print(s.name)
```

# Protected Variables (_variable)
Single underscore -> 'internal use' warning.

```python
class Student:
    def __init__(self):
        self._marks = 80

s = Student()
print(s._marks)
```

# Private Variables (__variable)
Double underscore -> hidden from outside.

```python
class Student:
    def __init__(self):
        self.__marks = 90

s = Student()
```

# Access Private Data using Methods (Getter and Setter)
Getter -> read value

```python
class Student:
    def __init__(self):
        self.__marks = 90
    def get_marks(self) -> int:
        return self.__marks

s = Student()
print(s.get_marks())
```

Setter -> update value safely

```python
class Student:
    def __init__(self):
        self.__marks = 90
    def set_marks(self, marks):
        if marks >= 0:
            self.__marks = marks
    def get_marks(self):
        return self.__marks

s = Student()
s.set_marks(85)
print(s.get_marks())
# Now data is controlled
```

# Real Example. Bank Account Protection.

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
    def get_balance(self):
        return self.__balance

acc = BankAccount(1000)
acc.deposit(500)
print(acc.get_balance())
```

# Inheritance (Reusing Code Between Classes)
## What is Inheritance
Inheritance = one class uses properties of another class.

```python
# Parent Class (Base Class)
class Animal:
    def eat(self):
        print("Animal eats")

# Child Class (Derived Class)
class Dog(Animal):
    def bark(self):
        print("Dog barks")

d = Dog()
d.eat()
d.bark()
```

```python
# Method Overriding (Very Important)
class Animal:
    def sound(self):
        print("Animal makes sound")
class Dog(Animal):
    def sound(self):
        print("Dog barks")

d = Dog()
d.sound()

f = Animal()
f.sound()
```

```python
# Using super() (Call Parent Method)
class Animal:
    def eat(self):
        print("Animal eats")
class Dog(Animal):
    def eat(self):
        super().eat()
        print("Dog eats bones")

d = Dog()
d.eat()
```

```python
# Multiple Inheritance
class A:
    pass
class B:
    pass
class C(A, B):
    pass
```

```python
# Real Example
class Employee:
    def __init__(self, name):
        self.name = name
    def show(self):
        print("Employee:", self.name)
class Manager(Employee):
    def manage(self):
        print("Manages team")

m = Manager('Deepak')
m.show()
m.manage()
```

# What is Polymorphism
Poly = many, Morphism = forms

## Example 1. Method Overriding (Runtime Polymorphism)

```python
class Animal:
    def sound(self):
        print("Animal makes sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

class Cat(Animal):
    def sound(self):
        print("Cat meows")

animals = [Dog(), Cat()]
for animal in animals:
    # Same method, Different outputs
    animal.sound()
```

## Example 2. Same Function, Different Objects

```python
class Car:
    def moves(self):
        print("Car moves")

class Plane:
    def moves(self):
        print("Plane flies")

def start_movement(vehicle):
    vehicle.moves()

# Function doesn't care about object type. This is polymorphism
start_movement(Car())
start_movement(Plane())
```

## Example 3. Built-in Polymorphism

```python
print(len("Python"))
print(len([1, 2, 3]))
# Same function works for string and list
```

**Types of Polymorphism**
1. Method Overiding Child class changes parent method.
2. Duch Typing(Python Style) "If it walks like a duck and quacks like a duck, it's a duck." Python doesn't check type - it checks behavior.

```python
class CreditCard:
    def pay(self, amount):
        print(f"Paid ${amount} using Credit Card.")

class UPI:
    def pay(self, amount):
        print(f"Paid ${amount} using UPI.")

def process_payment(method, amount):
    method.pay(amount)

process_payment(CreditCard(), 1000)
process_payment(UPI(), 300)
# same method pay()
# different implementations
```

**Task1: Bird class with Eagle and Penguin overiding fly()**

```python
class Bird:
    def fly(self):
        print("Bird flies")

class Eagle(Bird):
    def fly(self):
        print("Eagle flies high")

class Penguin(Bird):
    def fly(self):
        print("Penguin cannot fly")

birds = [Eagle(), Penguin()]
for b in birds:
    b.fly()
```

**Task 2. Shape class with Circle and Rectangle area()**

```python
class Shape:
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return 3.14 * self.radius * self.radius

class Rectangle(Shape):
    def __init__(self, length: int, width: int):
        self.length = length
        self.width = width
    
    def area(self):
        return self.length * self.width

shapes = [Circle(5), Rectangle(4,6)]
for s in shapes:
    print(f"Area: {s.area()}")
```

# Part 5: Abstraction  

Abstraction = show only important details and hide implementation.
- Hide how something works
- Show what it does  

**Real-Life Example**  

ATM Machine
- You press withdraw
- You don't know internal banking process  

Car driving
- You use steering and brake
- Engine mechanism is hidden  

**Why Abstraction is Important**  

- Reduce complexity
- Improve security
- Focus on essential features
- Cleaner system design

**Abstract Class (Blueprint Only)**  

Abstract class cannot create objects directly.

```python
from abc import ABC, abstractmethod
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
```

You cannot do: `s = Shape()` # Error  

Why? Because method not defined.

**Child Class Must Implement Abstract Method**

```python
from abc import ABC, abstractmethod
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
class Circle(Shape):
    def __init__(self, r):
        self.r = r
    def area(self):
        return 3.14*self.r*self.r

c = Circle(5)
print(c.area())
```

**Another Example. Payment System**

```python
from abc import ABC, abstractmethod
class Payment(ABC):
    @abstractmethod
    def pay(self, amount):
        pass
class UPI(Payment):
    def pay(self, amount):
        print(f"Paid: ${amount} using UPI")
class CreditCard(Payment):
    def pay(self, amount):
        print(f"Paid: ${amount} using CreditCard")

p = CreditCard()
p.pay(1000)
```

User only sees pay() - implementation hidden.

**Why You Need the Abstract Method**

The abstract method acts as a contract or blueprint that forces all child classes to implement certain methods. Here's why it's useful:

```python
class Shape:  # No abstract method
    def area(self):
        pass  # Empty method

class Circle(Shape):
    def __init__(self, r):
        self.r = r
    # Forgot to implement area() - no error!

circle = Circle(5)
print(circle.area())  # Returns None (silent failure)
```
